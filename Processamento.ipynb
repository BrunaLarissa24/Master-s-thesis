```python
def create_time_vector(dt, rec_len):
    """
    Create a time vector based on the given time step (dt) and record length (rec_len).

    Parameters:
        dt (float): Time step in seconds.
        rec_len (int): Length of the record in number of samples.

    Returns:
        numpy.ndarray: Time vector.
    """
    # Create time vector
    time_vector = np.arange(0, rec_len * dt, dt)
    return time_vector
```

#Criando o chirp

```python
# Define parameters
duration = 2  # Duration of the signal in seconds
sampling_rate = 1000000  # Sampling rate in Hz
t = np.linspace(0, duration, int(duration * sampling_rate))

# Generate chirp signal
f0 = 0.05  # Starting frequency in Hz
f1 = 800  # Ending frequency in Hz
t1 = 2  # Time at which f1 is achieved
signal = chirp(t, f0=f0, f1=f1, t1=t1, method='linear')

# Create figure and subplots
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(10, 8))
xi = 0
xf = 0.1
# Plot the chirp signal
ax1.plot(t, signal)
ax1.set_xlabel('Time (s)')
ax1.set_ylabel('Amplitude')
ax1.set_title('Chirp Signal')
#plt.xlim(xi,xf)
ax1.grid(True)

# Compute and plot the spectrogram
frequencies, times, Sxx = spectrogram(signal, fs=sampling_rate, nperseg = 100000)
ax2.pcolormesh(times, frequencies, 10 * np.log10(Sxx), shading='gouraud')
ax2.set_ylabel('Frequency (Hz)')
ax2.set_xlabel('Time (s)')
ax2.set_title('Spectrogram')
plt.ylim(0,1000)
#plt.xlim(xi,xf)
plt.tight_layout()
plt.show()


# Normalizar o sinal para valores inteiros entre -32768 e 32767 (16 bits PCM)
sinal_normalizado = np.int16(signal * 32767)

# Salvar o sinal em um arquivo .wav
write('sinal_senoidal.wav', sampling_rate,sinal_normalizado )
```
